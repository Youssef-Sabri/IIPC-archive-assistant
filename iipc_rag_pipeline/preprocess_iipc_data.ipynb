```python
# Read a CSV file using pandas
import pandas as pd
df = pd.read_csv('iipcm_extracted_content.csv')
df.head()
```

```python
import re

def clean_text(raw_text) -> str:
    if not isinstance(raw_text, str):
        return ""

    # Normalize line breaks
    text = raw_text.replace('\r', '\n')

    # Remove multiple newlines
    text = re.sub(r'\n+', '\n', text)

    # Remove bullet points and leading/trailing whitespace
    text = re.sub(r'^[\s•–\-]+\s*', '', text, flags=re.MULTILINE)
    text = re.sub(r'•|–|-', ' ', text)

    # Collapse multiple spaces
    text = re.sub(r'[ \t]+', ' ', text)

    # Strip leading/trailing spaces
    return text.strip()
```

```python
df['cleaned_text'] = df['full_text'].apply(clean_text)
```

```python
# Display original and cleaned text for the first row
print("=== ORIGINAL TEXT ===")
print(df['full_text'].iloc[0])
print("\n" + "="*50 + "\n")
print("=== CLEANED TEXT ===")
print(df['cleaned_text'].iloc[0])
print("\n" + "="*50 + "\n")
print(f"Character reduction: {len(df['full_text'].iloc[0]) - len(df['cleaned_text'].iloc[0])} chars")
print(f"Cleaning kept {len(df['cleaned_text'].iloc[0])/len(df['full_text'].iloc[0]):.1%} of original content")
```

```python
import time
import pandas as pd
from tqdm import tqdm
from google.generativeai import GenerativeModel, configure

# ✅ Best free model with high limits
MODEL_NAME = "gemini-2.5-flash-lite"  # Gemini 2.5 Flash-Lite

# Configure Gemini API
configure(api_key="add_your_api_key_here")  # Replace with your actual API key
model = GenerativeModel(MODEL_NAME)

# Cleaned column
cleaned_texts = []

# Process each row with 15 requests/minute (~4.1s delay)
for text in tqdm(df["full_text"], desc="Cleaning text with Gemini"):
    if pd.isna(text) or not text.strip():
        cleaned_texts.append("")
        continue

    try:
        response = model.generate_content(
            f"Make this text more meaningful and structured, but don't summarize: {text[:8000]}"
        )
        cleaned = response.text.strip()
    except Exception as e:
        print(f"⚠️ Error: {e}")
        cleaned = ""

    cleaned_texts.append(cleaned)
    time.sleep(4.1)  # Stay under 15 RPM limit

# Save to CSV
df["cleaned_text"] = cleaned_texts
df.to_csv("IIPCw_cleaned_text.csv", index=False)
print("✅ Saved cleaned text to IIPCw_cleaned_text.csv")
```

```python
import re
def clean_text(text):
    if pd.isna(text):
        return text
    # Remove lines starting with the specific AI-generated phrase
    text = re.sub(r"^Here's a more structured and meaningful.*$", '', text, flags=re.IGNORECASE | re.MULTILINE)

    # Remove lines starting with * or #
    text = re.sub(r'^[ \t]*[\*#][ \t]*', '', text, flags=re.MULTILINE)

    # Remove multiple newlines
    text = re.sub(r'\n{2,}', '\n', text)

    return text.strip()
```
